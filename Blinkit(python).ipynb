```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\DELL\Desktop\project 2\BlinkIT-Grocery-Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Item Weight'].fillna(df['Item Weight'].mean(), inplace=True)
```

```python
df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Outlet Type')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Outlet Type")
plt.show()
```
